```python
#3.Standardizing Data
# Check for outliers
print(df_new.describe())
def remove_outliers(df, column_name, lower_threshold=1.5, upper_threshold=1.5):
    Q1 = df_new[column_name].quantile(0.25)
    Q3 = df_new[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Remove outliers
    df_new2 = df_new[~((df_new[column_name] < Q1 - lower_threshold * IQR) | (df_new[column_name] > Q3 + upper_threshold * IQR))]

    return df_new2

# Call the function to remove outliers
df_new2 = remove_outliers(df_new, 'Electric Range')
df_new2 = remove_outliers(df_new, 'Model Year')
df_new2 = remove_outliers(df_new, 'Base MSRP')


print(df_new2)
```

```python
#normalize using to a specific range [0, 1] using MinMaxScaler
# Exclude non-numeric columns from normalization
numeric_columns = [ 'Postal_Code','Model_Year', 'Electric_Range', 'Legislative District', 'DOL Vehicle ID', '2020 Census Tract']
columns_to_normalize = [col for col in columns_to_normalize if col in numeric_columns]

min_max_scaler = MinMaxScaler()
df_new_process[columns_to_normalize] = min_max_scaler.fit_transform(df_new_process[columns_to_normalize])
```

```python
df_new_process.head()
```

```python
df_new_process['Electric Vehicle Type'].unique()
```

```python
# Replace 'Battery Electric Vehicle (BEV)' with 0, 'Plug-in Hybrid Electric Vehicle (PHEV)' with 1
df_new_process['Electric Vehicle Type'] = df_new_process['Electric Vehicle Type'].replace({'Battery Electric Vehicle (BEV)': 0, 'Plug-in Hybrid Electric Vehicle (PHEV)': 1})
```

```python
#saved to new data frame
df_new_process2 = df_new_process
df_new_process2.head()
```

#Data aggregation

```python
import numpy as np
# Display the original DataFrame
print("Original DataFrame:")
print(df_new_process2)

# Group by 'Model_Year' and 'Make' (company)
grouped_data = df_new_process2.groupby(['Model_Year', 'Make']).agg(Electric_Range=('Electric_Range', 'mean')).reset_index()

# Display the grouped data
print("\nGrouped Data by Model Year and Make (Electric_Range):")
print(grouped_data)

# Filter rows where how far a vehicle can travel purely on its electric charge.
filtered_to_group_data = grouped_data[grouped_data['Electric_Range'] > 0.0]

# Display the filtered data
print("\nFiltered Data (Electric_Range > 0):")
print(filtered_to_group_data)

# Create a pivot table to summarize data by describe how the electric range changes based on the model year and the companymake
pivot_table = pd.pivot_table(filtered_to_group_data, values='Electric_Range', index='Model_Year', columns='Make', fill_value=0)

# Display the pivot table
print("\nPivot Table:")
print(pivot_table)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import folium

# Descriptive statistics
description = df_new_process2.groupby('Model_Year').agg({'Electric Vehicle Type': 'unique', 'Electric_Range': 'mean'}).reset_index()

# Interactive plot using Plotly
#'Battery Electric Vehicle (BEV)' with 0, 'Plug-in Hybrid Electric Vehicle (PHEV)' with 1

fig = px.scatter(df_new_process2, x='Model_Year', y='Electric_Range', color='Electric Vehicle Type',
                 hover_data=['Electric Vehicle Type', 'Electric_Range'], log_x=True)
fig.update_layout(title='Electric Vehicle Range by Model Year',
                  xaxis_title='Model Year ',
                  yaxis_title='Electric Range (in hours)',
                  coloraxis_colorbar_title='Electric Vehicle Type')
fig.show()

# Descriptive statistics output
print(description)

```

```python
import pandas as pd
import plotly.express as px


# Filter the data for the year 2024
df_filtered = df_new_process2[df_new_process2['Model_Year'] == 2024]

# Group the filtered data by country, make, and model year, and count the occurrences
grouped_data = df_filtered.groupby(['Country', 'Make', 'Model_Year']).size().reset_index(name='Count')

# Filter for the most popular model year in each country and make
most_popular = grouped_data.groupby(['Country', 'Make']).apply(lambda x: x.nlargest(1, 'Count')).reset_index(drop=True)

# Create the interactive gang chart
fig = px.sunburst(most_popular, path=['Country', 'Make'], values='Count')

# Set the chart title
fig.update_layout(title='Popular Model Year in Each Country by company in year 2024')

# Display the chart
fig.show()
```
