```python
#3.Standardizing Data
# Check for outliers
print(df_new.describe())
def remove_outliers(df, column_name, lower_threshold=1.5, upper_threshold=1.5):
    Q1 = df_new[column_name].quantile(0.25)
    Q3 = df_new[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Remove outliers
    df_new2 = df_new[~((df_new[column_name] < Q1 - lower_threshold * IQR) | (df_new[column_name] > Q3 + upper_threshold * IQR))]

    return df_new2

# Call the function to remove outliers
df_new2 = remove_outliers(df_new, 'Electric Range')
df_new2 = remove_outliers(df_new, 'Model Year')
df_new2 = remove_outliers(df_new, 'Base MSRP')


print(df_new2)
```

```python
#rename columns
df_new2.rename(columns={'Postal Code':'Postal_Code'}, inplace=True)
df_new2.rename(columns={'Model Year':'Model_Year'}, inplace=True)
df_new2.rename(columns={'Electric Range':'Electric_Range'}, inplace=True)
display(df_new2.head())
```

```python
# Check for inconsistencies

# Correct inconsistencies
#ascindinng
def sort_data(data, columns, ascending=True):
    """
    Sort data in a DataFrame based on specified columns in ascending or descending order.

    Args:
        data (pandas.DataFrame): The DataFrame containing the data.
        columns (list): The columns to sort by.
        ascending (bool): Sort order (True for ascending, False for descending).

    Returns:
        pandas.DataFrame: The sorted DataFrame.
    """
    return data.sort_values(by=columns, ascending=ascending)
#use function
sort_columns = [ 'City', 'Model_Year', 'Electric_Range']

sorted_df = sort_data(df_new2, sort_columns, ascending=True)
display(sorted_df)
```
